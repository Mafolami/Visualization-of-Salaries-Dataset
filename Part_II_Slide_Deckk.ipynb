```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
sals = pd.read_csv("C:\\Users\\SURFACE LAPTOP 4\\Downloads\\Visualization\\Project Template\\salaries.csv")
```

## (Visualization 1)

> This violinplot shows that TotalPayBenefits increase every new year. The following barchart will suggest whether increase in the number of emploees or job titles every year affects TotalPayBenefits!

```python
#Check TotalPayBenefits by year
plt.figure(figsize=[12,8])
sns.violinplot(data = sals, x = 'Year', y = 'TotalPayBenefits')
plt.title('TotalPayBenefits by year')
plt.xlabel('Year')
plt.ylabel('TotalPayBenefits')
plt.show()
```

## (Visualization 2)

> There are over 2000 unique jobs in the dataset but the follwoing 10 are the most common jobs

```python
plt.figure(figsize=[10,6])
toptitle = sals['JobTitle'].value_counts().head(10)
toptitle.plot(kind='bar')
plt.title('Top Ten most common jobs')
plt.xlabel('Job Titles')
plt.ylabel ('count of job titles')
```

## (Visualization 3)

The chart below shows the relationship between TotalPay and TotalPayBenefits. As expected, the total amount of money earned by an employee correlates positively with the TotalPayBenefits. That is, the higher an employees TotalPay, the higher also is their TotalPayBenefits.

```python
g = sns.FacetGrid(data = sals, hue = 'Year', height = 6)
g.map(plt.scatter, 'TotalPay', 'TotalPayBenefits',  marker=",", alpha=0.5, s=5)
g.add_legend()
plt.title('TotalPay vs TotalPayBenefits')
```
